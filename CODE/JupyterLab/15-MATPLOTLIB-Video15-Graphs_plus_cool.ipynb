```python
# chargement des librairies
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# dataset
iris = load_iris()
# print(iris)

# données
x = iris.data
# les targets classe 0 1 ou2
y = iris.target
names = list(iris.target_names)

print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f'Il y a {np.unique(y).size} classes')
print('')
print(x)
print(y)
```

## (5) Graphique de CLASSIFICATION : plt.scatter

```python
# on fait passer en abscisse et ordonnées 2 des 4 variables 
# de notre problème : par ex première variable et 2ième variable
# pour toutes les lignes
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
# colorier ces points suivant la nature de leur classes y = 0 1 ou 2
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.xlabel('Longueur sépal')
plt.ylabel('Largeur du sépal')
plt.show()
```

```python
# limite : onne peut représenter que 2 variables sur les 4
```

```python
# astuce : utiliser le paramètre alpha pour contrôler la transparence
# de chacun des points du graph.
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5)
plt.xlabel('Longueur sépal')
plt.ylabel('Largeur du sépal')
plt.show()
```

```python
# astuce : utiliser le paramètre s qui contrôle la taille des points
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=100)
plt.xlabel('Longueur sépal')
plt.ylabel('Largeur du sépal')
plt.show()
```

```python
# astuce : controler la taille en fonction d'une variable du problème
# ex : taille en fonction de la variable 3 x[:, 2]
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:, 2]*100)
plt.xlabel('Longueur sépal')
plt.ylabel('Largeur du sépal')
plt.show()
```

```python
#  plus les points sont gros plus le pétal est long et inversement
```

## (4-2) surface : plot.surface()

```python
# visualiser des surfaces et fonctions maths modélisées
# créer une fonction f, générer par lambda générateur de fonction
# anonyme dans python
f = lambda x, y : np.sin(x) + np.cos(x+y)
# pour observer dans un graphique 3D :
# créer 2 vecteurs numpy allant de 0 à 5
# créer une sorte de grille : np.meshgrid à partir de 2 axes
X = np.linspace(0, 5, 100) # 100 points
Y = np.linspace(0, 5, 100)
# recréer X et Y
X, Y = np.meshgrid(X, Y)

# il reste à injecter X, Y dans la fonction f  dans variable Z :
Z = f(X, Y)
```

```python
# dimension de Z : 100 lignes 100 colonnes
Z.shape
```

```python
# %matplotlib
ax = plt.axes(projection='3d')
# préciser la couleur : cmap
ax.plot_surface(X, Y, Z, cmap='plasma')
```

## (2) Contour plots : contour()

```python
# Repartir de la fonction f pour la visualiser en 2D  vue du dessus
# visualiser des surfaces et fonctions maths modélisées
# créer une fonction f, générer par lambda générateur de fonction
# anonyme dans python
f = lambda x, y : np.sin(x) + np.cos(x+y)
# pour observer dans un graphique 3D :
# créer 2 vecteurs numpy allant de 0 à 5
# créer une sorte de grille : np.meshgrid à partir de 2 axes
X = np.linspace(0, 5, 100) # 100 points
Y = np.linspace(0, 5, 100)
# recréer X et Y
X, Y = np.meshgrid(X, Y)

# il reste à injecter X, Y dans la fonction f  dans variable Z :
Z = f(X, Y)

# visualisation en 2D en vue du dessus
plt.contour(X, Y, Z)
plt.show()
```

```python
# visualisation en 2D en vue du dessus
# pour être plus visible : modifier le nombre de niveaux à 20
# augmente la précision del'affichage
plt.contour(X, Y, Z, 20)
plt.show()
```

```python
# visualisation en 2D en vue du dessus
# pour être plus visible : modifier le nombre de niveaux à 20
# augmente la précision del'affichage
# encore plus lisible : modifier la couleur
plt.contour(X, Y, Z, 20, colors='black') # tous les positifs en continus,
# les négatifs en discontinus
# on voit les maximum ou minimum si fonction coû
plt.show()
```

### avec une colormap : countourf(... cmap=)

```python
# visualisation en 2D en vue du dessus
# pour être plus visible : modifier le nombre de niveaux à 20
# augmente la précision del'affichage
# encore plus lisible : modifier la couleur avec colormap
plt.contourf(X, Y, Z, 20, cmap='RdGy') # tous les positifs en continus,
# les négatifs en discontinus
# on voit les maximum ou minimum si fonction coût
plt.show()
```

```python
# visualisation en 2D en vue du dessus
# pour être plus visible : modifier le nombre de niveaux à 20
# augmente la précision del'affichage
# encore plus lisible : modifier la couleur avec colormap
# encore plus : avec légende
plt.contourf(X, Y, Z, 20, cmap='RdGy') # tous les positifs en continus,
# les négatifs en discontinus
# on voit les maximum ou minimum si fonction coût
plt.colorbar()
plt.show()
```

## CORRECTION EXERCICE VIDEO 14

```python
# Un dictionnaires des données des 100 expériences
dataset={f'experience{i}': np.random.randn(100) for i in range(4)}
dataset

# créer une fonction graphique qui imprime sur 1 même figure le dataset
#  4 expériences ==> 4 graphiques
def graphique(dataset):
    # taille du dictionnaire
    n = len(dataset)
    # figure agrandie dans lequel on va placer les 4 graphiques    
    plt.figure(figsize=(12,30))
    # boucle sur le dictionnaire
    for exp_cle, i in zip(dataset.keys(), range(1,n+1)):
        # Graphique i
        # n figure à l'emplacement i de 1 à 4       
        plt.subplot(n, 1, i)
        # alimente les ordonnées de l'expérience         
        plt.plot(dataset[exp_cle])
        # joute du titre = clé de l'expérience du dictionnaire
        plt.title(exp_cle)
    # tracer la figure    
    plt.show()
```

```python
graphique(dataset)
```

```python
# Un dictionnaires des données des 100 expériences à 6 classes
dataset_6={f'experience{i}': np.random.randn(100) for i in range(6)}
graphique(dataset_6) # 6 graphiques
```

```python
# Un dictionnaires des données des 100 expériences à 6 classes
# mais on collecte 3 variables
dataset_6_3={f'experience{i}': np.random.randn(100, 3) for i in range(6)}
graphique(dataset_6_3) # 6 graphiques
# chacune des 3 variables est affichée ur les graphiques
```
