## 2. NETTOYER LES DONNEES

### Eliminer des variables de la df : drop()

```python
# pour éliminer des variables inutiles du dataframe : drop()
# avec le nom des différentes colonnes qu'on veut supprimer selon axe des colonne = 1
data = data.drop(['name','sibsp', 'parch', 'ticket','fare', 'cabin', 
                  'embarked', 'boat', 'body', 'home.dest'], axis=1)
```

## TRAITER DONNEES MANQUANTES

## OPTION 1  - fillna() : remplacer par valeur par défaut

## OPTION 2 - dropna() : supprimer des lignes

```python
# pour titanic pour âge supprimer les lignes ==> option 2
data = data.dropna(axis=0)
```

```python
# dimensions ? 1046 lignes et 4 colonnes
data.shape
```

```python
# cela change les statistiques : 40 %  sur 1046 ont survécus
# au lieu de 28% sur 1349
data.describe()
```

### value_counts()

```python
# trouver parmi les passagers ceux en 1er ou 2nde ou 3ieme classe
data['pclass'].value_counts()
```

```python
# premier graphique nbr passagers par classe
df_passager_par_classe = data['pclass'].value_counts()
df_passager_par_classe.plot.bar()
plt.show()
```

```python
# passager par ages sous forme histogramme
df_passager_par_age=data['age']
df_passager_par_age.hist()
plt.show()
```

### groupby()

```python
#  regrouper les gens selon le sexe
data.groupby(['sex']).mean()
# 70%des femmes ont survécus et 30% seulement pour les hommes
```

```python
# combinaison des regroupements : par age et par classe dans laquelle
# ils voyageaient
data.groupby(['sex','pclass']).mean()
# sur les 70 % de femmes qui ont survécus 37% voyageaient en 1ere classe
```

## INDEXING : ILOC et LOC

### indexing : iloc()

```python
# iloc permet de faire de l'indexing sur un dataframe
#  en utilisant les index
# car on ne peut pas le faire directement : sinon erreur
data[0:2, 0:2]
```

```python
data.iloc[0:2, 0:2]
```

## indexing : loc()

```python
# loc permet de faire de l'indexing sur un dataframe
#  en utilisant les colonnes
data.loc[0:2, 'age']
```

```python
#  pour indiquer plus d'une colonne ==> on passe une liste de colonne
data.loc[0:2, ['age', 'sex']]
```

```python
#  pour avoir toutes les colonnes : préciser '':'
data.iloc[0:2,:]
```

## EXERCICE VIDEO 17

```python
data_3 = pd.read_excel('titanic.xls')
# modifier la colonne age pour créer 4 catégorie 0-20, 20-30, 30-40, >40
# technique de feature engineering :  faire des analyses avec plus de sens
# transforme la colonne age en mettant 0 si 0-20, 1 si 20-30, 2 si 30-40
# 4 si >40  via masque sur colonne agepuis on compte le nombre 
data_3.loc[data_3['age']<=20, 'age'] = 0
data_3.loc[ (data_3['age']>20) & (data_3['age']<=30), 'age' ] = 1
data_3.loc[(data_3['age']>30) & (data_3['age']<=40), 'age'] = 2
data_3.loc[data_3['age']>40, 'age'] = 3
data_3['age']                               
                               
```
